# Exploratory Data Analysis - Employee Attrition

This notebook explores the IBM HR Analytics Employee Attrition dataset: overview, missing values, class balance, correlations, distributions, and attrition drivers.

```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

from src.data.data_loader import load_raw_data

df = load_raw_data(project_root=PROJECT_ROOT)
df.head()
```

## Dataset overview

```python
print("Shape:", df.shape)
print("\nInfo:")
df.info()
print("\nDescribe:")
df.describe()
```

## Missing values analysis

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
if len(missing) > 0:
    print("Columns with missing values:")
    print(missing)
    sns.barplot(x=missing.values, y=missing.index)
    plt.title("Missing values per column")
    plt.tight_layout()
    plt.show()
else:
    print("No missing values.")
```

## Class imbalance (Attrition)

```python
attrition_counts = df['Attrition'].value_counts()
print("Attrition distribution:")
print(attrition_counts)
print("\nProportions:")
print(df['Attrition'].value_counts(normalize=True))

fig, ax = plt.subplots(figsize=(5, 4))
ax.bar(['No (0)', 'Yes (1)'], [attrition_counts.get(0, 0), attrition_counts.get(1, 0)], color=['steelblue', 'coral'])
ax.set_ylabel('Count')
ax.set_title('Attrition distribution')
plt.tight_layout()
plt.show()
```

## Correlation heatmap (numerical)

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[num_cols].corr()
fig, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(corr, annot=False, cmap='RdBu_r', center=0, ax=ax, fmt='.1f')
ax.set_title('Correlation heatmap (numerical features)')
plt.tight_layout()
plt.show()
```

## Feature distributions

```python
key_numerical = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']
df[key_numerical].hist(bins=20, figsize=(12, 8), layout=(2, 3))
plt.suptitle('Feature distributions')
plt.tight_layout()
plt.show()
```

## Attrition drivers – visualizations

```python
# Attrition by Department
fig, ax = plt.subplots(figsize=(8, 4))
pd.crosstab(df['Department'], df['Attrition']).plot(kind='bar', ax=ax)
ax.set_title('Attrition by Department')
ax.legend(['No', 'Yes'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Attrition by Overtime
fig, ax = plt.subplots(figsize=(5, 4))
pd.crosstab(df['OverTime'], df['Attrition']).plot(kind='bar', ax=ax)
ax.set_title('Attrition by OverTime')
ax.legend(['No', 'Yes'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Attrition by income (box plot)
fig, ax = plt.subplots(figsize=(6, 4))
df.boxplot(column='MonthlyIncome', by='Attrition', ax=ax)
ax.set_title('Monthly income by Attrition')
ax.set_xlabel('Attrition (0=No, 1=Yes)')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# Attrition by tenure (YearsAtCompany)
fig, ax = plt.subplots(figsize=(6, 4))
df.boxplot(column='YearsAtCompany', by='Attrition', ax=ax)
ax.set_title('Years at company by Attrition')
ax.set_xlabel('Attrition (0=No, 1=Yes)')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# Attrition vs Job Satisfaction
fig, ax = plt.subplots(figsize=(6, 4))
pd.crosstab(df['JobSatisfaction'], df['Attrition'], normalize='index').plot(kind='bar', ax=ax)
ax.set_title('Attrition rate by Job Satisfaction')
ax.legend(['No', 'Yes'])
ax.set_ylabel('Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Attrition vs Work Life Balance
fig, ax = plt.subplots(figsize=(6, 4))
pd.crosstab(df['WorkLifeBalance'], df['Attrition'], normalize='index').plot(kind='bar', ax=ax)
ax.set_title('Attrition rate by Work Life Balance')
ax.legend(['No', 'Yes'])
ax.set_ylabel('Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
